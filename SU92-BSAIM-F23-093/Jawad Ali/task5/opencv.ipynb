```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python

img = cv2.imread(r'C:\Users\Saeed Ahmad\Desktop\PIA taks\task5\pic2.webp')

plt.imshow(img)  
plt.title("Original Image")
plt.axis("on") 
plt.show() 
```

```python
convert_colors = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(convert_colors)
plt.title("Gray Picture") 
plt.axis("off")  
plt.show()
```

```python
convert_colors = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(convert_colors)
plt.title("RGB Picture") 
plt.axis("off")  
plt.show()
```

```python
resized = cv2.resize(img, (300, 300))
plt.imshow(resized)
plt.title("Resized Image")
plt.axis("on")
plt.show()

```

```python

image = cv2.imread(r"C:\Users\Saeed Ahmad\Desktop\PIA taks\task5\pic2.webp")  
(h, w) = image.shape[:2]

center = (w // 2, h // 2)

M = cv2.getRotationMatrix2D(center, 45, 1.0)

rotated = cv2.warpAffine(image, M, (w, h))
plt.imshow(rotated)
plt.title("Rotated Image")
plt.axis("on")
plt.show()

```

```python

flipped = cv2.flip(image, 0)
plt.imshow(flipped) 
plt.title("Edges")
plt.axis("off")  
plt.show()
```

```python

image = cv2.imread(r"C:\Users\Saeed Ahmad\Desktop\PIA taks\task5\pic2.webp", 0)  

edges = cv2.Canny(image, 100, 200)

plt.imshow(edges, cmap='gray') 
plt.title("Edges")
plt.axis("off")  
plt.show()
```

```python

image = cv2.imread(r"C:\Users\Saeed Ahmad\Desktop\PIA taks\task5\pic2.webp")  

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
_, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
output = np.zeros_like(image)

cv2.drawContours(output, contours, -1, (0, 255, 0), 2)

output_rgb = cv2.cvtColor(output, cv2.COLOR_BGR2RGB)

plt.imshow(output_rgb)
plt.title("Contours")
plt.axis("off")
plt.show()
```
